```python
import matplotlib.pyplot as plt
```

```python
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
plt.scatter(x,y, color='green', label='bitches', marker='D', s=y)
plt.xlabel('Total Number of Bitches')
plt.ylabel('Average Fare ($)')
plt.title('Average Fare per Bitch')
plt.legend()
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create scatter plot
plt.scatter(y_axis,x_axis, color='red', label='Chicago')
plt.xlim(0,50)
plt.gca().invert_yaxis()
#annotations
plt.ylabel('Month')
plt.xlabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
#create scatter plot
plt.scatter(x_axis,y_axis, color='red', label='Chicago', s=y_axis)
plt.ylim(0,50)

#annotations
plt.ylabel('Month')
plt.xlabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
#we need to make the bubbles bigger so that we can actually see them... multiply data points by 3
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
#create scatter plot
plt.scatter(x_axis,y_axis, color='red', label='Chicago', s=y_axis_larger)
plt.ylim(0,50)

#annotations
plt.ylabel('Month')
plt.xlabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
#we need to make the bubbles bigger so that we can actually see them... multiply data points by 3
   
#create scatter plot
plt.scatter(x_axis,y_axis, color='red', label='Chicago', s = [i * 3 for i in y_axis])
plt.ylim(0,50)

#annotations
plt.ylabel('Month')
plt.xlabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis], color='skyblue', edgecolors='black', linewidth=2, label='Boston')
ax.invert_yaxis()
ax.set_xlabel('Fare($)')
ax.set_ylabel('Month')
ax.set_title('Monthly PyBer Fares')
ax.set_xlim(0,50)
ax.legend()
```
