```python
import matplotlib.pyplot as plt
```

```python
categories = ['pants','shirts','outerwear','pajamas','underwear','shoes']
```

```python
revenue = [25,15,5,20,10,25]
```

```python
plt.pie(revenue,labels=categories)
plt.show()
```

```python
plt.pie(revenue,labels=categories,autopct='%1.1f%%')
plt.show()
```

```python
explode = [0,.1,0,0,0,0]
```

```python
plt.pie(revenue,labels=categories,autopct='%1.1f%%',explode=explode)
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.pie(y_axis, labels=x_axis)
plt.show
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
```

```python
#explode July out
plt.pie(y_axis, labels=x_axis, explode=explode_values)
plt.show
```

```python
# add values to each slice using autopct='%1.1f%%'
# to format one decimal place = .1f%
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%1.1f%%')
plt.show
```

```python
#since the shits is crowded, expand plt.subplots(figsize=(8,8))
plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%1.1f%%')

plt.show
```

```python
#mpl uses standard color cycle 
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.subplots(figsize=(8,8))
plt.pie(y_axis, labels=x_axis, explode=explode_values, autopct='%1.1f%%', colors=colors)

plt.show
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show
```

```python
#add percentages
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis, autopct='%1.1f%%')
plt.show
```

```python
#explode two biggest months
explode = [0,0,.2,0,0,0,.2,0,0,0,0,0]
fig, ax = plt.subplots()
ax.pie(y_axis, labels=x_axis, autopct='%1.1f%%',explode=explode)
plt.show
```

```python
#increase fig size to 8x8
explode = [0,0,.2,0,0,0,.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, autopct='%1.1f%%',explode=explode)
plt.show
```

```python
#add a shadow
explode = [0,0,.2,0,0,0,.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, autopct='%1.1f%%',explode=explode, shadow=True)
plt.show
```

```python
#rotate so Jan is on top
explode = [0,0,.2,0,0,0,.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, autopct='%1.1f%%',explode=explode, shadow=True, startangle=90)
plt.show
```

```python
#reverse order
explode = [0,0,.2,0,0,0,.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, autopct='%1.1f%%',explode=explode, shadow=True, startangle=90, counterclock=False)
plt.show
```

```python
#change colors
explode = [0,0,.2,0,0,0,.2,0,0,0,0,0]
fig, ax = plt.subplots(figsize=(8,8))
ax.pie(y_axis, labels=x_axis, autopct='%1.1f%%',explode=explode, shadow=True, startangle=90, counterclock=False, colors=colors)
plt.show
```
