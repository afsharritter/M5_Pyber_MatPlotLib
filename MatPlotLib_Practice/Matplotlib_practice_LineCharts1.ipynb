```python
import matplotlib.pyplot as plt
```

```python
savings = [0,100,150,200,300,500,300,350,500,600,750,1000]
months = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings in Dollars')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
plt.xlabel('Months')
plt.ylabel('SUDS Scores')
plt.title('Psychometrics')
plt.show
```

```python
# Create the plot with ax.plt()

# plt.subplots() function returns a tuple that contains a figure and axis (or axes) (both objects)
# fig, ax = plt.subplots() unpacks the tuple into variables fig and ax
fig, ax = plt.subplots()
# fig references the figure
# ax references the axes of the figure, allowing us to annotate if we'd like

# ax.plot(x_axis, y_axis) plots the data from the x and y axes
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt
fig = plt.figure()
ax = fig.add_subplot()
#now add x_axis and y_axis data to ax.plt()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt
fig = plt.figure()
ax = fig.add_subplot()
#now add x_axis and y_axis data to ax.plt()
ax.plot(x_axis, y_axis)
#add labels and titles
ax.set_xlabel('months')
ax.set_ylabel('Fare')
```
